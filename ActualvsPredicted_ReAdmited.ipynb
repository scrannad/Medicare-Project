```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Import and read the csv file
df = pd.read_csv("numeric_only.csv")
```

```python
# Drop the target column from the input data
input_data = df.drop('Hospital_Readmit_Count', axis=1)
```

```python
# Select the target column
target_data = df['Hospital_Readmit_Count']

X_train, X_test, y_train, y_test = train_test_split(input_data, target_data, test_size=0.2, random_state=42)
```

```python
# Scale the input features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

input_features = X_train_scaled.shape[1]
model = tf.keras.Sequential()
```

```python
# Define your model layers here
layer1 = 5
layer2 = 10

# First hidden layer
model.add(tf.keras.layers.Dense(units=layer1, input_dim=input_features, activation='relu'))

# Second hidden layer
model.add(tf.keras.layers.Dense(units=layer2, activation='relu'))

# Output layer
model.add(tf.keras.layers.Dense(units=1))  # Assuming you're predicting a single value

# Compile the model with appropriate optimizer and loss function
model.compile(optimizer='adam', loss='mse')
```

```python
# Train the model
model.fit(X_train_scaled, y_train, epochs=10, batch_size=32)
```

```python
# Evaluate the model on the test set
loss = model.evaluate(X_test_scaled, y_test)
print('Test Loss:', loss)
```

```python
# Make predictions
predictions = model.predict(X_test_scaled)
```

```python
# Print predictions and actual values
for i in range(len(predictions)):
    print('Predicted:', predictions[i][0], 'Actual:', y_test.iloc[i])
```
