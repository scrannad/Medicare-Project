```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from google.colab import drive
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from sklearn.linear_model import LinearRegression

from sklearn.linear_model import LogisticRegression
```

```python
file_path = 'https://raw.githubusercontent.com/scrannad/Medicare-Project/Janice_branch_set/numeric_only.csv'
df = pd.read_csv(file_path)
df.head(25)
```

```python

#Check out data types
df.info()
```

```python
print(df.columns)
```

```python
correlation_matrix = df.corr()
correlation_with_target = correlation_matrix['Hospital_Readmit_Rate%'].abs().sort_values(ascending=False)
print(correlation_with_target)
```

```python
relevant_features = correlation_with_target[correlation_with_target >= 0.2].index.tolist()
```

```python
relevant_features.append('Average_Age')
```

```python
selected_features_data = df[relevant_features]
```

```python
features = df[['Average_Age']] # Exclude the target variable
target = df['Hospital_Readmit_Rate%']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()  # Instantiate the chosen regression model
model.fit(X_train, y_train)  # Train the model
```

```python
y_pred = model.predict(X_test)  # Make predictions on the test set
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

 low R2 value suggests that there may be other important factors influencing the readmission rate that are not captured by age alone. 

```python
import matplotlib.pyplot as plt

plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, y_pred, color='red', label='Predicted')
plt.xlabel('Average Age')
plt.ylabel('Hospital Readmission Rate')
plt.title('Predicted vs Actual Hospital Readmission Rate')
plt.legend()
plt.show()
plt.savefig('scatter_plot.png') 
```
